```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.stattools import acovf, acf, pacf, pacf_ols, pacf_yw
import warnings
warnings.filterwarnings('ignore')
from pandas.plotting import lag_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Non-stationary data
df1 = pd.read_csv('Data/airline_passengers.csv',index_col=0,parse_dates=True)
df1.head()
```

```python
df1.index.freq = 'MS'
```

```python
# Stationary data
df2 = pd.read_csv('Data/DailyTotalFemaleBirths.csv',index_col=0,parse_dates=True)
df2.head()
```

```python
df2.index.freq = 'D'
```

```python
df = pd.DataFrame({'a': [13,5,17,9,12]})
df.head()
```

```python
acf(df['a'])
```

```python
pacf_yw(df['a'],nlags=4,method='unbiased')
```

```python
pacf_yw(df['a'],nlags=4,method='mle')
```

```python
pacf_ols(df['a'],nlags=4)
```

```python
lag_plot(df1['Thousands of Passengers']);
```

```python
lag_plot(df2['Births']);
```

```python
df1.plot(figsize=(8,4));
```

```python
plot_acf(df1,lags=40);
```

```python
df2.plot(figsize=(8,4));
```

```python
plot_pacf(df1,lags=40,method='ywunbiased');
```

```python
plot_pacf(df1,lags=40,method='ywmle');
```

```python
plot_pacf(df1,lags=40,method='ols');
```

```python
plot_pacf(df2,lags=40,method='ywunbiased',title='Partial Autocorrelation - Daily Female Births');
```

```python
plot_pacf(df2,lags=40,method='ywmle',title='Partial Autocorrelation - Daily Female Births');
```

```python
plot_pacf(df2,lags=40,method='ols',title='Partial Autocorrelation - Daily Female Births');
```
