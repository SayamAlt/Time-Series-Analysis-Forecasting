```python
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('Data/airline_passengers.csv',index_col=0,parse_dates=True)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.plot(figsize=(12,5));
```

```python
results = seasonal_decompose(df['Thousands of Passengers'],model='multiplicative')
results
```

```python
results.trend
```

```python
results.seasonal
```

```python
results.resid
```

```python
results.nobs
```

```python
results.observed
```
