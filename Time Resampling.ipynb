```python
# custom function
def first_entry(entry):
    if len(entry):
        return entry[0]
```

```python
def last_entry(records):
    if len(records):
        return records[-1]
```

```python
df.resample(rule='A').apply(first_entry)
```

```python
df.resample(rule='A').apply(last_entry)
```

```python
df['Close'].resample(rule='A').mean().plot.bar(title='Starbucks Yearly Average Closing Price');
```

```python
df['Close'].resample(rule='M').max().plot.bar(figsize=(16,6),title='Starbucks Monthly Max Closing Price',color=['#1f77b4']);
```
