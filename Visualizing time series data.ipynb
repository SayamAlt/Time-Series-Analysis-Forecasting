```python
import pandas as pd
from matplotlib import dates
```

```python
df = pd.read_csv("Data/starbucks.csv",index_col='Date',parse_dates=True)
df.head()
```

```python
df.plot();
```

```python
df['Close'].plot();
```

```python
df['Volume'].plot();
```

```python
ax = df['Close'].plot(figsize=(12,6),title='Close Price')
ax.set_xlabel('Date')
ax.set_ylabel('Close')
ax.autoscale(enable=True,axis='both',tight=True);
```

```python
ax = df['Close'].plot(figsize=(12,6),title='Close Price')
ax.set_xlabel('Date')
ax.set_ylabel('Close')
ax.autoscale(enable=True,axis='x',tight=True);
```

```python
df['Close']['2018-09-17':'2019-04-02'].plot(figsize=(12,4));
```

```python
df['Close'].plot(figsize=(12,4),xlim=['2017-08-26','2019-01-01']);
```

```python
df['Close']['2017-08-26':'2019-01-01'].plot(figsize=(12,5),ylim=[40,70],ls='--',c='green');
```

```python
ax = df['Close'].plot(xlim=['2018-03-01','2018-05-30'],ylim=[55,59],figsize=(12,5))
ax.set(xlabel='');

ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n\n\n\n%b'))

ax.xaxis.grid(True)
ax.yaxis.grid(True);
```
