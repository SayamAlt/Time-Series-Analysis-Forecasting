```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# y = mx + b + noise
m = 4
b = 3
x = np.linspace(0,50,500)
```

```python
np.random.seed(101)
noise = np.random.normal(loc=0,scale=5,size=(500,))
```

```python
noise.shape, x.shape
```

```python
y = m * x + b + noise
```

```python
plt.plot(x,y,'--');
```

```python
model = Sequential()
model.add(Dense(units=8,input_dim=1,activation='relu'))
model.add(Dense(units=16,activation='relu'))
model.add(Dense(units=1,activation='linear'))
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
r = model.fit(x,y,epochs=100)
```

```python
plt.plot(range(100),r.history['loss']);
```

```python
x_test = np.linspace(0,50,500)

predictions = model.predict(x_test)
predictions
```

```python
plt.plot(x,y,'*')
plt.plot(x_test,predictions,'r');
```

```python
print("MAE:", mean_absolute_error(y,predictions))
print("MSE:", mean_squared_error(y,predictions))
print("RMSE:", np.sqrt(mean_squared_error(y,predictions)))
```
