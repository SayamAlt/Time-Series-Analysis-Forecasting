```python
import pandas as pd
```

```python
df = pd.read_csv("Data/starbucks.csv",index_col='Date',parse_dates=True)
df.head()
```

```python
df['Close'].plot(figsize=(12,5));
```

```python
df.rolling(window=7).mean().head(10)
```

```python
df.rolling(window=10).mean().head(20)
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(7).mean()['Close'].plot(title='Weekly Rolling Window (SMA7 - Simple Moving Averages with Window Size=7)');
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(30).mean()['Close'].plot(title='Monthly Rolling Window (SMA30 - Simple Moving Averages with Window Size=30)');
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=60).mean()['Close'].plot(title='Rolling Window with size 60 days (SMA60)');
```

```python
df['Close: Simple Moving Average 7 days'] = df.rolling(window=7).mean()['Close']
df[['Close','Close: Simple Moving Average 7 days']].plot(figsize=(12,5));
```

## Expanding

```python
df['Close'].expanding().mean().plot(figsize=(12,5),title='Closing Price Expanding Window'); # Converse towards the average value of the entire column, "Close" in this case
```
